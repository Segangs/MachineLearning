```python
month_se = pd.Series(['1월', '2월', '3월', '4월'])
income_se = pd.Series([9500, 6200, 6050, 7000])
expenses_se = pd.Series([5040, 2350, 2300, 4800])
df = pd.DataFrame({'월': month_se, '수입': income_se, '지출': expenses_se})
print(df['월'])
print(df['수입'][0])
```

```python
df = pd.read_csv("./csv/vehicle_prod.csv", index_col=0) # 인덱스 컬럼을 따로 생성하지 않고 국가이름 컬럼으로 지정
print(df) # 전부 문자열로 넘어옴
```

```python
df['2007'] #문자열로 넘어오기 때문에 2007이 아닌 '2007'이다.
```

```python
list1 = df.columns.to_list() # 파이썬에서 쓸수있게 리스트형으로 받아온다.
print(list1)
```

```python
np1 = df.columns.to_numpy()
print(np1) #넘파이 타입으로 받아온다.(컴마가 없다)
```

```python
df['total'] = df.sum(axis=1) # 컬럼을 새로 만들고 가로로 더하시오.
print(df)
```

```python
df['mean'] = df[['2007', '2008', '2009', '2010', '2011']].mean(axis=1) # 컬럼을 새로 만들고 가로 평균값을 구하시오. (total 컬럼은 빼고)
print(df)
```

```python
df.drop('2007', inplace=True, axis=1) # 2007컬럼을 삭제하시오.
```
