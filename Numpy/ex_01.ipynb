```python
a = np.arange(0,8)
print(a)

def train_test_split(numbers):
    np.random.shuffle(numbers)
    divisor = int(len(numbers) * 0.8)
    train = numbers[:divisor]
    test = numbers[divisor:]
    return (train, test)
x = np.arange(1, 51)
(train, test) = train_test_split(x)
print(f"train: {train}, \nTest : {test}")
```
