```python
import tensorflow as tf
from keras.src import datasets, layers, models
```

```python
(train_images, train_labels), (test_images, test_labels) = datasets.mnist.load_data()
```

```python
train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))
```

```python
print(train_images.shape)
```

```python
print(test_images.shape)
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input((28,28,1)))

model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu'))
model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))

# 한번 더 한다 (컨벌루션 + 맥스풀링 쌍으로)
model.add(tf.keras.layers.Conv2D(filters=16, kernel_size=(3,3), activation='relu'))
model.add(tf.keras.layers.Conv2D(filters=16, kernel_size=(3,3), activation='relu'))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(16, activation='relu'))

model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])  # 컴파일 한다.
(train_images, test_images) = (train_images / 255.0, test_images / 255.0) # 정규화 한다.
```

```python
model.fit(train_images, train_labels, epochs=20)
```

```python
model.save("CNN_MNIST_20260629-1540.keras")
```
