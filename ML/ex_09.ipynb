# MNIST 필기체 숫자 인식 (신경망 이용)

```python
import keras
mnist = keras.datasets.mnist
(X_train, y_train),(X_test, y_test) = mnist.load_data()
(X_train, X_test) = (X_train / 255.0, X_test / 255.0)
```

```python
model = keras.Sequential()
model.add(keras.layers.Input(shape=(28, 28)))
model.add(keras.layers.Flatten())  #평탄화
model.add(keras.layers.Dense(512, activation='relu'))  #relu말고 다른것 써도 된다.
model.add(keras.layers.Dropout(0.2))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) # 0 ~ 9 Sparse Categorical Crossentroy
model.summary()
```

```python
model.fit(X_train, y_train, epochs=20)
model.evaluate(X_test, y_test)
```

```python
model.save('MNIST_20260629_11-35.keras') # 모델을 저장한다.
# pc자원을 많이 먹고 오래 걸리기 때문에, colab으로 돌려놓은 다음에 저장한 모델을 불러오는게 낫다.
```
