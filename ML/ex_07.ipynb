# MNIST 필기체 숫자 데이터 세트

```python
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist.load_data()
# print(mnist)
# print(mnist[0])
print(mnist[0][0])

data=mnist[0][0]
print(data.shape)
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train.shape
y_train.shape
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Input(shape=(28*28,)))
model.add(tf.keras.layers.Dense(units=512, activation='relu', name='Hidden1'))
model.add(tf.keras.layers.Dense(units=256, activation='relu', name='Hidden2'))
model.add(tf.keras.layers.Dense(units=10, activation='softmax', name='outputlayer')) # 10개는 결과값
```

```python
model.summary()
```

```python
X_images = X_train.reshape((60000, 28*28))
type((X_images[0] / 255).dtype)
```

```python
X_images_norm = (X_images / 255).astype('float32')
```

```python
X_images_test = X_test.reshape((10000, 28*28))
X_images_test_norm = (X_images_test / 255).astype('float32')
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_images_norm, y_train, epochs=30) # 학습시킨다. norm을 붙인다. accuracy가 1에 가까울수록 학습이 잘되는것. 
```

```python
model.save("MNIST_260626_15-30.keras") # 잘된 모델을 저장한다. (하고나서 )
model.evaluate(X_images_test_norm, y_test)
```
