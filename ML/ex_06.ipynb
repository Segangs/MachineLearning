```python
import numpy as np
import tensorflow as tf
from matplotlib import pyplot as plt
import keras
```

```python
# 데이터를 학습 데이터와 테스트 데이터로 나눈다.
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

data_size= x_train.shape[0]
batch_size= 12 # 배치 크기
selected = np.random.choice(data_size, batch_size)

print(selected)
x_batch= x_train[selected]
y_batch= y_train[selected]
```

## keras 이용

```python
mlp = keras.models.Sequential()
input_layer = keras.layers.Input(shape=(2,))

# 모델 생성
dense1 = keras.layers.Dense(32, activation='sigmoid')
dense2 = keras.layers.Dense(1, activation='sigmoid')
```

### 적층하기

```python
mlp.add(input_layer)
mlp.add(dense1)
mlp.add(dense2)
mlp.summary()
```

### 입력값 넣기

```python
X = tf.constant([[0, 0], [0, 1], [1, 0], [1, 1]])
y = tf.constant([0, 1, 1, 0])
```

### 합치기(모델 컴파일)

```python
mlp.compile(loss='mse', optimizer='sgd')
mlp.fit(X, y, epochs=1000)
```
