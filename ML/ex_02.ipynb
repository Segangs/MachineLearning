# 첫 번째 머신러닝 프로그램 (선형 회귀)

```python
from sklearn import linear_model  # sciket-learn 설치후 import (pip install scikit-learn)
import matplotlib.pyplot as plt
import numpy as np
```

```python
reg = linear_model.LinearRegression()  # 클래스를 가져와라
```

### 값을 넣는다.

```python
X = [[0], [1], [2]]  # 무조건 2차원으로 넣어야 함
Y = [3, 4, 5]
```

### 학습을 시킨다.

```python
reg.fit(X, Y)  # 학습을 시킨다
```

```python
reg.coef_  # 직선의 기울기
```

### 함수가 나왔다. y=ax+b

```python
reg.intercept_  # 직선의 y-절편.. 함수가 나왔다. y=1x+3
```

### 예측을 시켜본다.

```python
reg.predict([[5]])  # 5일때 나올 값을 예측해 봐라
```

```python
plt.scatter(X, Y, c="black")
plt.plot(X, reg.predict(X), c="blue", linewidth=2)
plt.show()
```

# 예제 실습

```python
exreg = linear_model.LinearRegression()  # 클래스를 가져와라
```

```python
X = [[174], [152], [138], [128], [186]]  # 무조건 2차원으로 넣어야 함
Y = [71, 55, 46, 38, 88]
```

```python
exreg.fit(X, Y)  # 학습을 시킨다
```

```python
exreg.intercept_  # 직선의 y절편
```

```python
plt.scatter(X, Y, c="black")
plt.scatter([165], [69.09], c="red")
plt.plot(X, exreg.predict(X), c="blue", linewidth=3)
plt.show()
```

```python
x1 = np.arange(120, 190)
y1 = 0.82 * x1 - 68.02

plt.scatter(X, Y, c="black")
plt.scatter([165], [67.30], c="blue", marker='*')
plt.plot(x1, y1, c="cyan", linewidth=3)
plt.show()
```
