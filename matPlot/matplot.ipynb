# 데이터 시각화 도구 : matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x = np.arange(-10, 11)
y = np.square(x)

plt.plot(x, y)  # 이어서 봄
# plt.stem(x,y) #디스크립트

plt.axis([0, 100, 0, 100])  # 가로로 넓게 퍼트리지 않고(왜곡) 원래 1:1 비율의 그래프로 본다.
plt.show()
```

## 2𝑥, (1/3)𝑥2 + 5, −𝑥2 − 5의 세 함수 그리기

```python
x = np.arange(-20, 21)  # -20에서 20 사이의 수를 1의 간격으로 생성
y1 = x * 2  # 2*x를 원소로 가지는 y1 함수
y2 = (1 / 3) * np.square(x) + 5  # (1/3)*× ** 2 + 5를 원소로 가지는 y2 함수
y3 = -(x ** 2) - 5  # -x ** 2 - 5를 원소로 가지는 y3 함수
plt.plot(x, y1, 'g--')  #초록 점실선
plt.plot(x, y2, 'b-*')  # 파란색 별
plt.plot(x, y3, 'r.:')  # 빨강 점선
plt.axis([-30, 30, -30, 30])  # 그림을 그릴 영역을 지정함
plt.show()
```

## sin, cos 그리기 / 이미지 저장

```python
x = np.linspace(0, 2 * np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, 'b--', x, y2, 'r-.')
plt.savefig("sine-cosine.jpg")  # 이미지로 저장하기
plt.show()
```

## 제목과 레이블, 스타일

```python
x = np.linspace(0, 2 * np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, 'b--', label="Sine wave")
plt.plot(x, y2, 'r--', label="Cosine wave")
plt.legend()  # 범례 쓰기

plt.title("TITLE")
plt.xlabel("time")
plt.ylabel("mag")

plt.show()
```

## 다양한 종류의 그래프를 한 화면에 : figure, axes

```python
fig, ax = plt.subplots(2, 2) # 2행 2열의 그림을 그리는 공간을 만든다

X = np.random.randn(100) #random normal , 정규 분포를 가지는 데이터
Y = np.random.randn(100)
ax[0, 0].scatter(X, Y)  # 산점도 그림

X = np.arange(10)   # 0에서 9 사이의 연속값
Y = np.random.uniform(1, 10, 10)    # 균일분포값 생성
ax[0, 1].bar(X, Y)  # 막대 차트 그림

X = np.linspace(0, 10, 100)
Y = np.cos(X)
ax[1, 0].plot(X, Y)  # 실선으로 함수를 그림

Z = np.random.uniform(0, 1, (5, 5))
ax[1, 1].imshow(Z)  # 분포를 2D 이미지(imshow)로 그림
```

## subplot() 격자 이용하여 개별 그래프 크기 조정

```python
fig, ax = plt.subplots(2,3)

grid = plt.GridSpec(2, 3, wspace=0.4, hspace=0.3)

X = np.random.randn(100)
Y = np.random.randn(100)

plt.subplot(grid[0,0]).scatter(X, Y)
X = np.arange(10)
Y = np.random.uniform(1, 10, 10)
plt.subplot(grid[0, 1:]).bar(X, Y)

X = np.linspace(0, 10, 100)
Y = np.cos(X)
plt.subplot(grid[1, :2]).plot(X, Y)

Z = np.random.uniform(0, 1, (5,5))
plt.subplot(grid[1,2]).imshow(Z)
```

```python
x = np.arange(3)
years = ['2010','2011', '2012']
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1681]

plt.bar(x, domestic, width=0.25)
plt.bar(x + 0.3, foreign, width=0.25)
plt.xticks(x, years)
```

```python
data = [5,4,6,11]
clist = ['cyan', 'gray', 'orange', 'red']
explode = [.06, .07, .08, .09]
plt.pie(data, autopct = '%.2f%%', colors = clist, labels=clist, explode=explode) #explode는 원형 그래프를 쪼갠다
plt.show()
```

## 히스토그램

```python
heights = np.array([175, 165, 164, 164, 171, 165, 160, 169, 164, 159, 163, 167, 163,172, 159, 160, 156, 162, 166, 162, 158, 167, 160, 161, 156, 172, 168, 165, 165, 177])
plt.hist(heights, bins=6)   # 히스토그램 bins는 구간의 수
plt.xlabel("height")
plt.ylabel("frequency")

plt.hist(heights, bins=6, label='cumulative=True', cumulative=True)     # 누적된 자료값의 합
plt.hist(heights, bins=6, label='cumulative=False', cumulative=False)
plt.legend(loc='upper left')
plt.show()
```
